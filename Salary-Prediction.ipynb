```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
filename = "C:\\Users\\rroha\\OneDrive\\Desktop\\Machine-Learning-A-Z-Codes-Datasets\\Machine Learning A-Z\\Part 1 - Data Preprocessing\\Section 2 -------------------- Part 1 - Data Preprocessing --------------------\\Python\\Data.csv"
dataset = pd.read_csv(filename)
X= dataset.iloc[:,:3].to_numpy()
y= dataset.iloc[:,-1].to_numpy()
print(dataset)
imputer = SimpleImputer(missing_values = np.nan,strategy ='mean')
imputer.fit(X[:,[1,2]])
X[:,[1,2]] = imputer.transform(X[:,[1,2]])
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])],remainder = 'passthrough')
X = ct.fit_transform(X)
le = LabelEncoder()
y = le.fit_transform(y)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
sc = StandardScaler()
X_train[:,[3,4]] = sc.fit_transform(X_train[:,[3,4]])
X_test[:, [3,4]] = sc.transform(X_test[:,[3,4]])
```
